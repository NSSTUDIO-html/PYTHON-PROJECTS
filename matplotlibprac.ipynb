```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)              # Line plot
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, 'g--o')  # Green dashed line with circles
plt.title("Styled Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 22]

plt.plot(x, y1, label='Data 1', color='orange', marker='o')
plt.plot(x, y2, label='Data 2', color='green', linestyle='--', marker='s')

plt.title("Multiple Lines")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()  # Show labels
plt.grid(True)
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [5, 7, 3, 8]

plt.bar(x, y, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 25, 20]
colors = ['red', 'yellow', 'pink', 'brown']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Fruit Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='purple')
plt.title("Simple Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
ages = [22, 55, 62, 45, 21, 22, 34, 42, 42, 26, 70, 67, 90, 18, 21, 30, 25, 22, 40, 36]

plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  # 2 rows, 1 column, plot 1
plt.plot(x, y1, label='sin(x)', color='blue')
plt.title("Sine and Cosine")

plt.subplot(2, 1, 2)  # 2 rows, 1 column, plot 2
plt.plot(x, y2, label='cos(x)', color='green')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, linestyle='--', marker='o', color='orange', linewidth=2, markersize=8)
plt.title("Styled Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label='Squares', color='blue')
plt.plot(x, y2, label='Line', color='green')
plt.legend(loc='upper left')
plt.title("Legend Example")
plt.show()
```

```python
import matplotlib.pyplot as plt

students = ['Sasank', 'Teja', 'Yash', 'Ravi', 'Anu']
marks = [85, 78, 92, 65, 88]

plt.bar(students, marks, color='skyblue')
plt.title('Student Marks Comparison')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

plt.plot(x, y1, label='x^2', marker='o')
plt.plot(x, y2, label='x^3', marker='s')
plt.title("Power Functions")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
labels = ['Rent', 'Groceries', 'Bills', 'Savings']
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Monthly Budget')
plt.axis('equal')
plt.show()
```

```python
subjects = ['Math', 'Science', 'English', 'History']
scores = [90, 75, 88, 70]

plt.barh(subjects, scores, color='purple')
plt.title("Subject Scores")
plt.xlabel("Scores")
plt.ylabel("Subjects")
plt.show()
```
